```
import pandas
import matplotlib.pyplot as plt
import numpy as np

file_name = "https://raw.githubusercontent.com/FS-Jorda/Introduction-to-Python/master/united_states_covid19_cases_and_deaths_by_state.csv"
case_data = pandas.read_csv(file_name)

case_data.plot(x = "State/Territory", y = "Case Rate per 100000",kind = "bar")
plt.tick_params(axis ='x', rotation = 90, labelsize=6)
                        
plt.show()

file_name = "https://raw.githubusercontent.com/mlepinski/Python-Worksheets/master/Matplotlib/Data/covid_approval_toplines.csv"
approval_data = pandas.read_csv(file_name)

approval_data.plot(x = "party", y = "approve_estimate",kind = "scatter")
plt.tick_params(axis ='x', rotation = 90, labelsize=10)
             
plt.show()

approval_data.plot(x = "party", y = "disapprove_estimate",kind = "scatter")
plt.tick_params(axis ='x', rotation = 90, labelsize=10)
             
plt.show()

file_name = "https://raw.githubusercontent.com/mlepinski/Python-Worksheets/master/Matplotlib/Data/covid_concern_toplines.csv"
concern_data = pandas.read_csv(file_name)

concern_data.plot(x = "subject", y = "very_estimate",kind = "scatter")
plt.tick_params(axis ='x', rotation = 90, labelsize=10)
             
plt.show()
```
